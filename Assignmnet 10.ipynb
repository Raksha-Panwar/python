```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df=pd.read_csv('Footballer.csv')
df.head()
```

### What is the distribution  of preferred foot?


```python
plt.figure(figsize=(5,5))

mode_PreferredFoot=df['Preferred Foot'].mode().values[0]
# print(mode_PreferredFoot)
df['Preferred Foot'].replace(np.nan,mode_PreferredFoot,inplace=True)
# print(df['Preferred Foot'].isnull().sum())

sns.countplot(data=df, x='Preferred Foot')


ax = plt.gca()
# Add text labels above the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.show()
```

### How can we visualize the age distribution of players?

```python
plt.figure(figsize=(15,12))

sns.countplot(data=df, x='Age')

ax = plt.gca()

# Add text labels above the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')
plt.show()
```

### What is the international reputation of left and right foot?

```python
plt.figure(figsize=(8,5))

median_InternationalReputation = df['International Reputation'].median()
# print(median_InternationalReputation)

df['International Reputation'].replace(np.nan, median_InternationalReputation, inplace=True)
# print(df['International Reputation'].isnull().sum())

sns.countplot(data=df, x='International Reputation', hue=df['Preferred Foot'])

ax = plt.gca()
# Add text labels above the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.show()
```

### What is the different work rate of the Players?

```python
mode_WorkRate=df['Work Rate'].mode().values[0]
# print(mode_WorkRate)
df['Work Rate'].replace(np.nan,mode_WorkRate,inplace=True)
# print(df['Work Rate'].isnull().sum())


plt.figure(figsize=(12,10))

sns.countplot(data=df,x='Work Rate')


ax = plt.gca()
# Add text labels above the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.show()
```

### Top 10 countries with most number of players

```python
df['Nationality'].value_counts()[:10].plot(kind='bar')

ax = plt.gca()
# Add text labels above the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')
    
plt.show()
```

### What is the work rate of players segregated by preferred foot?

```python
plt.figure(figsize=(12,10))
sns.countplot(data=df,x='Work Rate', hue='Preferred Foot')


ax = plt.gca()
# Add text labels above the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.show()
```

### Number of players at different positions? 



```python
mode_Position=df['Position'].mode().values[0]
# print(mode_Position)
df['Position'].replace(np.nan,mode_Position,inplace=True)
# print(df['Position'].isnull().sum())

df['Position'].value_counts()
```

```python
plt.figure(figsize = (20, 8))
ax = sns.countplot(data = df, x= 'Position', palette = 'Dark2_r')
ax.set_xlabel(xlabel = 'Positions in Football', fontsize = 16)
ax.set_ylabel(ylabel = 'Number of Players', fontsize = 16)
ax.set_title(label = 'Count of Players with Positions', fontsize = 20)   
plt.show()
```

### Age distribution in top countries

```python
top_countries = df['Nationality'].value_counts()[:10].index
print(top_countries)

plt.figure(figsize=(15,12))

df_country = df.loc[df['Nationality'].isin(top_countries)]
plt.figure(1 , figsize = (10,6))
sns.boxplot(x = 'Nationality' , y = 'Age' , data = df_country, palette='Accent')

plt.show()
```
