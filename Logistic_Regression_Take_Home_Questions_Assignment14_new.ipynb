### 5. Build classification model and present it's classification report ?

```python
df.sample(3)
```

```python
df['numeral_class'] = [1 if df.loc[i,'Class'] == 'H' else 2 if df.loc[i,'Class'] == 'M' else 3 for i in range(len(df))]
```

```python
dummy_data = df.drop(['Class'] , axis=1)
new_df = pd.get_dummies(dummy_data , drop_first=True)  #0 in dummy datafield means category which is not separately created

new_df
```

```python
X = new_df.drop(['numeral_class'] , axis=1)
Y = new_df['numeral_class']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X)

X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size = 0.25 , random_state=7)   #7 saga newton-cg #
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(solver='newton-cg')
lr.fit(X_train , Y_train)
print(lr.score(X_train, Y_train))
print(lr.score(X_test, Y_test))
```

or

```python
predictions = lr.predict(X_test)

from sklearn.metrics import accuracy_score, classification_report
accuracy_score(Y_test, predictions)
```

```python
print(classification_report(Y_test, predictions))
```
