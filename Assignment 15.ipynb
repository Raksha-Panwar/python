### Q.4 Convert the target variable â€œChance of Admitâ€ to categorical having values 0 and 1,such that :
<br>Students having the â€œChance of Admitâ€ value &gt; 0.80, are assigned value 1, and
<br>Students having the â€œChance of Admitâ€ value &lt; 0.80, are assigned value 0
<br>Where 0: Low chance of Admission and 1: High chance of admission

```python
df['numeric_ChanceofAdmit'] = [1 if df.loc[i,'Chance of Admit '] >= 0.80 else 0 for i in range(len(df))]
df['numeric_ChanceofAdmit'].unique()
```

### Q.5 Build a Decision Tree classifier, to predict whether a student has a low or high chance of admission to a chosen university. Perform Hyperparameter Tuning to improve the accuracy of the model.

```python
X = df.drop(['numeric_ChanceofAdmit','Serial No.'] , axis=1)
Y = df['numeric_ChanceofAdmit']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size =0.2, random_state =3)
```

```python
# from sklearn.linear_model import LogisticRegression

# reg = LogisticRegression(solver='newton-cg')
# reg.fit(X_train, Y_train)
# print(reg.score(X_train, Y_train))
# print(reg.score(X_test, Y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier()
dtc.fit(X_train, Y_train)
print(dtc.score(X_train, Y_train))
print(dtc.score(X_test, Y_test))
```
